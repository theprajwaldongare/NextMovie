```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Bollywood Dataset

```python
df = pd.read_csv("movieData.csv")
df
```

```python
df = df[['Title','Review','Genre']]
df
```

```python
df.shape
```

```python
len(df['Title'].unique())
```

```python
features = df.columns.to_list()
for i in features:
    print(df[i].isna().sum())
```

```python
df['Genre'].isna()
```

```python
df = df.drop_duplicates(subset=['Title'])
```

```python
df
```

```python
df['Review'] = df['Review'].str.replace('See full summary', '', regex=False).str.strip()
```

```python
df.rename(columns={"Title":"title","Review":"desc","Genre":"genre"},inplace=True)
```

```python
df
```

```python
df.to_csv('Datasets/bollywood.csv',index=False)
```
