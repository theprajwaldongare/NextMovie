```python
df = pd.read_csv("IMDB_10000.csv")
df
```

```python
df = df.dropna()
df.shape
```

```python
features = df.columns.to_list()
print(features)
```

```python
titleFeature = df['title']
descFeature = df['desc']
genreFeature = df['genre'].str.replace(","," ")
# genreFeature
descFeature
```

```python
uniqueFeatures = ['year','certificate','genre']
genre = []
for i in uniqueFeatures:
    if i=='genre':
        genre = df[i].unique()
    print(i)
    print(df[i].unique())
    print("\n")
```

#### Genre based Only

```python
# genre = genre.tolist()
# print(genre)
uniqueGenre = set()
for i in genre:
    Glist = i.split(",")
    for j in Glist:
        if j[0]==" " or j[len(j)-1]==" ":
            removeSpace = j.split(" ")
            for u in removeSpace:
                if u!=" ":
                    uniqueGenre.add(u)
        else:
            uniqueGenre.add(j)
print(uniqueGenre)
print(len(uniqueGenre))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

vec = vectorizer.fit_transform(genreFeature)
print(vectorizer.get_feature_names_out())

search = "drama thriller"
srchVec = vectorizer.transform([search])

# print(vec.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similar = cosine_similarity(srchVec,vec).flatten()
print(similar)

topK = 5
topIndex = np.argsort(similar)[::-1][:topK]
print(topIndex)
```

```python
for i in topIndex:
    print(df['title'].iloc[i])
    print("Cosine Value: ", similar[i])
    print("\n")
```

```python
descFeature
```

```python
vectorizer = CountVectorizer()

decvec = vectorizer.fit_transform(descFeature)
print(vectorizer.get_feature_names_out())

# searchDec = "Action based new cool fiction story"
searchDec = "horror"
srchDecVec = vectorizer.transform([searchDec])

# print(vec.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarDec = cosine_similarity(srchDecVec,decvec).flatten()
print(similarDec)

topK = 5
topIndexDec = np.argsort(similarDec)[::-1][:topK]
print(topIndexDec)
```

```python
for i in topIndexDec:
    print(df['title'].iloc[i])
    print("Cosine Value: ", similarDec[i])
    print(df['desc'].iloc[i])
    print("\n")
```
