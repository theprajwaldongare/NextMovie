#### Desc & Genre (Bollywood)

```python
descGenreFeature = genreFeature.fillna("") + descFeature.fillna("")
descGenreFeature
```

```python
print(len(descGenreFeature))
print(len(df))
```

```python
# vectorizer = CountVectorizer()
# vectorizer = CountVectorizer(stop_words="english")
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english")

decvec = vectorizer.fit_transform(descGenreFeature)
print(vectorizer.get_feature_names_out())

# searchDec = "Action based new cool fiction story"
# searchDec = "horror"
# searchDec = df["desc"].iloc[100]
searchDec = "serial killer murder investigation"

srchDecVec = vectorizer.transform([searchDec])

# print(vec.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarDec = cosine_similarity(srchDecVec,decvec).flatten()
print(similarDec)

topK = 5
topIndexDec = np.argsort(similarDec)[::-1][:topK]
print(topIndexDec)
```

```python
print("Search keyword: ",searchDec)
print("\n")
for i in topIndexDec:
    print(df['title'].iloc[i])
    print("Cosine Value: ", similarDec[i])
    print(df['desc'].iloc[i])
    print("\n")
```

#### Desc & Genre (Hollywood)

```python
hdf = pd.read_csv("Datasets/hollywood.csv")
hdf
```

```python
hdf.shape
```

```python
hdf.dropna(subset=['desc'], inplace=True)
```

```python
htitleFeature = hdf['title']
hdescFeature = hdf['desc']
hgenreFeature = hdf['genre'].str.replace(","," ")
# genreFeature
hdescFeature
```

```python
hdescGenreFeature = hgenreFeature.fillna("") + hdescFeature.fillna("")
hdescGenreFeature
```

```python
print(len(hdescGenreFeature))
print(len(hdf))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
hvectorizer = TfidfVectorizer(stop_words="english")

hdecvec = hvectorizer.fit_transform(hdescGenreFeature)
print(hvectorizer.get_feature_names_out())

# hsearchDec = "Action based new cool fiction story"
# hsearchDec = "horror"
# hsearchDec = hdf["desc"].iloc[100]
hsearchDec = "serial killer murder investigation"

hsrchDecVec = hvectorizer.transform([hsearchDec])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

hsimilarDec = cosine_similarity(hsrchDecVec,hdecvec).flatten()
print(hsimilarDec)

topK = 5
htopIndexDec = np.argsort(hsimilarDec)[::-1][:topK]
print(htopIndexDec)
```

```python
print("Search keyword: ",hsearchDec)
print("\n")
for i in htopIndexDec:
    print(hdf['title'].iloc[i])
    print("Cosine Value: ", hsimilarDec[i])
    print(hdf['desc'].iloc[i])
    print("\n")
```
